```python
from ultralytics import YOLO
import cv2
import tkinter as tk
from tkinter import messagebox
import platform
import threading
from playsound import playsound

# Caminho para o modelo
model_path = "yolov5nu.pt"
model = YOLO(model_path)

# Vídeo de entrada
video = "video14.mp4"
cap = cv2.VideoCapture(video)

# Parâmetros
consecutive_person_frames = 0
threshold_frames = 3
confidence_threshold = 0.50

# Controle do beep
keep_beeping = False

# Função que gera beep contínuo
def beep_loop():
    global keep_beeping
    while keep_beeping:
        playsound("alarm2.mp3")

# Função para mostrar alerta com beep contínuo
def show_person_alert(frames_detected):
    global keep_beeping

    # Inicia o beep contínuo em thread separada
    keep_beeping = True
    threading.Thread(target=beep_loop, daemon=True).start()

    # Mostra popup
    root = tk.Tk()
    root.withdraw()
    messagebox.showwarning(
        "PERSON ALERT",
        f"ALERT: PERSON or KID detected for {frames_detected} consecutive frames!"
    )

    # Ao fechar a janela, para beep
    keep_beeping = False
    root.destroy()

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Detecção
    results = model(frame)
    annotated = results[0].plot()

    person_detected_this_frame = False

    for box_idx in range(len(results[0].boxes)):
        conf = results[0].boxes.conf[box_idx].item()
        label = results[0].names[int(results[0].boxes.cls[box_idx])]

        if label.lower() in ['person', 'kid', 'cat',
                             'dog', 'horse', 'sheep', 'cow'] and conf > confidence_threshold:
            person_detected_this_frame = True
            break

    # Atualiza contador
    if person_detected_this_frame:
        consecutive_person_frames += 1
    else:
        consecutive_person_frames = 0

    # Se atingiu limiar
    if consecutive_person_frames >= threshold_frames:
        show_person_alert(consecutive_person_frames)
        break

    # Mostra vídeo com boxes
    cv2.imshow("Person/Kid Detection", annotated)
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

# Libera recursos
cap.release()
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import cv2
import tkinter as tk
from tkinter import messagebox
import threading
from playsound import playsound

# Load model
model = YOLO("yolov5nu.pt")

# Load video
cap = cv2.VideoCapture("video15.mp4")

# Parameters
threshold_frames = 3
confidence_threshold = 0.85
consecutive_frames = 0
keep_beeping = False

# Target labels to detect
target_labels = ['person', 'kid', 'cat', 'dog', 'horse', 'sheep', 'cow']

# Function: play alarm sound in loop
def beep_loop():
    global keep_beeping
    while keep_beeping:
        playsound("alarm2.mp3")

# Function: show popup alert with sound
def show_alert(label, frames_detected):
    global keep_beeping
    keep_beeping = True
    threading.Thread(target=beep_loop, daemon=True).start()

    root = tk.Tk()
    root.withdraw()
    messagebox.showwarning(
        "DETECTION ALERT",
        f"ALERT: {label.upper()} detected for {frames_detected} consecutive frames!"
    )

    keep_beeping = False
    root.destroy()

# Main loop
while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Run detection
    results = model(frame)
    annotated = results[0].plot()

    # Check for target labels
    detected = False
    detected_label = ""

    for box in results[0].boxes:
        conf = box.conf.item()
        label = results[0].names[int(box.cls)]

        if label.lower() in target_labels and conf > confidence_threshold:
            detected = True
            detected_label = label
            break

    # Update counter
    consecutive_frames = consecutive_frames + 1 if detected else 0

    # Trigger alert
    if consecutive_frames >= threshold_frames:
        show_alert(detected_label, consecutive_frames)
        break

    # Show video with detections
    cv2.imshow("Detection", annotated)
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

# Cleanup
cap.release()
cv2.destroyAllWindows()
```
